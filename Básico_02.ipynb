```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

#Saturação
c = contraste \\
b = brilho \\
$\hat{f}(x,y) = f(x,y) \cdot c + b$

```
#SATURAÇÃO
def sat(f, b, c):
  f_ajuste = np.zeros(f0.shape)

  for x in range(0, f0.shape[0]):
    for y in range(0, f0.shape[1]):
      if (f[x,y]*c + b) > 255:
        f_ajuste[x,y] = 255
      elif (f[x,y]*c + b) < 0:
        f_ajuste[x,y] = 0
      else:
        f_ajuste[x,y]*c + b
        
        
  return plt.imshow(f_ajuste, cmap='gray')

sat(f1, 150, 1)
```

```
#alpha blend

fn0 = ((f0 - f0.min())/(f0.max() - f0.min()))
fn1 = ((f1 - f1.min())/(f1.max() - f1.min()))

alfa = 0.5
h = fn0*alfa + fn1*(1-alfa)


plt.imshow(h, cmap='gray')
```

#Correção do gamma

$g(x,y) = f(x,y)^{\gamma}$

##Histograma

```
h = np.zeros(256)
n = np.arange(256)

for x in range(0, f0.shape[0]):
  for y in range(0, f0.shape[1]):
      v = f0[x,y]
      h[v] = h[v] + 1
      
plt.bar(n,h)
```

```
p = h/h.sum()
c = p.cumsum()
transformacao = ((c*(2**8-1)).astype('uint16')) 

image_equalizada = transformacao[f0]

plt.figure()
plt.subplot(1,2,1)
plt.imshow(f0, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(image_equalizada, cmap='gray')
```

```
plt.bar(n, c)
```

```
h_eq = np.zeros(256)

for x in range(0, image_equalizada.shape[0]):
  for y in range(0, image_equalizada.shape[1]):
      v = image_equalizada[x,y]
      h_eq[v] = h_eq[v] + 1
      
plt.figure()
plt.subplot(2,1,1)
plt.bar(n, h)
plt.subplot(2,1,2)
plt.bar(n, h_eq)
```

```
histograma, bins = np.histogram(f0.ravel(), 256)
plt.bar(n, histograma)
```
